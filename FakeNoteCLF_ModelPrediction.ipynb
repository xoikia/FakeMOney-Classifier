# Importing necessary libraries 

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image


#this magic function helps to visulaize the graph at the ouput 
%matplotlib inline
```

## **Creating the Label**

The label list contains the name of the classes we are predicting , the index position are based on what our inage datagenerator class had given while building the model

```python
label=['200_fake', '200_real', '500_fake', '500_real']
```

# Creating  function  to call the model to predict

This function will load the model and the image and the display it 
and finally predict the class and print out italicized text

```python
def fake_classifier(model,loc):
    

  '''
    DOC STRING:
    the fake_classifier function take the model path  and the location of the
    image path as input
  '''
    
    
    #using  image to load the image
  img=image.load_img(loc,target_size=(224,224))

    #plots the image
  plt.imshow(img)

    #converting the image into numpy array for predicition
  img=np.asarray(img)

    #expansion of dimensions
  img=np.expand_dims(img,axis=0)
    
    #loading the model to predict
  CLF_model=load_model(model)

    #predicitng the output and storing in a variable called out
  out=CLF_model.predict(img)

    #label is the list we have deined above 
  '''
    DOC STRING: np.argmax(out) gives the  position/index of the out variable 
     which havehighest value and that value is used to access the value from 
     that label list


    eg out=[0,0,1,0]
    np.argmax(out) will give index position 2 since in index position 2 of out 
    has highest value  ,now this index value 2 will be used to acces the  label,
    returning the value stored in index  2 of label ,whixh is the label
    ''' 

  print(label[np.argmax(out)])
```
